```python
import numpy as np
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv("data/heart_attack_prediction_dataset.csv")
df = pd.DataFrame(dataset).drop(columns=["Patient ID", "Continent", "Hemisphere"])
```

```python
df["Blood Pressure (systolic)"]= df["Blood Pressure"].str.replace('\d{1,3}\/', '', regex=True).astype(int)
df["Blood Pressure (diatolic)"] = df["Blood Pressure"].str.replace('\/\d{1,3}', '', regex=True).astype(int)
df = df.drop(columns=["Blood Pressure"]);
```

```python
df.columns, df.dtypes
```

```python
# convert to numerical data only
# extract non-numerical features
features = ["Sex", "Diet", "Country"]

transformer = ColumnTransformer([
    ('one_hot', OneHotEncoder(), features)
], remainder="passthrough")
```

```python
X = df.drop(columns=["Heart Attack Risk"])
y = df["Heart Attack Risk"]
```

```python
transformed_X = transformer.fit_transform(X)
pd.DataFrame(transformed_X)
```

```python
# split data into train & test
X_train, x_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)
```

```python
# train a model
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
# evaluate model
model.score(x_test, y_test)
```

```python
#export model
import pickle

pickle.dump(model, open("kaggle_melbourne_housing_prices_model.pkl", "wb"))
```
