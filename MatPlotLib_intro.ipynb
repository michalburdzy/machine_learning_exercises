```python
fig, ((ax0, ax1, ax2)) = plt.subplots(nrows=1, ncols=3, figsize=(10,3))
```

```python
# Subplots option 2
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(10,5))
axes[0,0].plot(x, x/2);
axes[0,1].scatter(np.random.random(10),np.random.random(10));
axes[1,0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
axes[1,1].hist(np.random.randn(100));
```

```python
over_50 = heart_disease[heart_disease.age > 50]
over_50
```

```python
# Pyplot method
over_50.plot(kind="scatter", x='age', y='chol', c='target', figsize=(10,6))
```

```python
# OO method mixed with pyplot
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter", x='age', y='chol', c='target', ax=ax);

# set x axis limits
ax.set_xlim([45, 90])
```

```python
# OO method from scratch
fig, ax = plt.subplots(figsize=(10,6))

# plot the data
scatter = ax.scatter(x=over_50.age, y=over_50.chol, c=over_50.target)

# Customize
ax.set(title="Heart disease, age and Cholesterol levels",
      xlabel="age",
       ylabel="cholesterol",
      );

# add a legend
ax.legend(*scatter.legend_elements(), title="Heart disease appearance")

# Add a horizontal line
ax.axhline(over_50.chol.mean(), linestyle="dotted");
```

```python
heart_disease["chance"] = 1 / ((heart_disease.chol * heart_disease.thal ) + heart_disease.age) * 10000

heart_disease[heart_disease.target == 0]
```

```python

# subplot of chol, age thalach
fig, (ax1, ax2) = plt.subplots(nrows=2,
                              ncols=1,figsize = (10,10))
scatter = ax1.scatter(x=over_50.age, y=over_50.chol, c=over_50.target);
ax1.set(title="Heart disease for age & cholesterol level", xlabel='age', ylabel="chol")
ax1.set_xlim(45, 85)
ax1.axhline(y=over_50.chol.mean(), linestyle="--");
ax1.legend(*scatter.legend_elements(), title="Target")

scatter2 = ax2.scatter( x=over_50.age, y=over_50.thalach, c=over_50.chance);
ax2.set_xlim(45, 85)
ax2.set(title="Heart disease for age & thalach",xlabel='age', ylabel="max heart rate");
ax2.legend(*scatter2.legend_elements(), title="Predicted heart disease chance");

fig.suptitle("Heart Disease Analysis", fontsize=12, fontweight='bold');
```

### Customising x and y limitations

```python
plt.style.use('seaborn-v0_8')

fig, (ax1, ax2) = plt.subplots(nrows=2,
                              ncols=1,figsize = (10,10))
scatter = ax1.scatter(x=over_50.age, y=over_50.chol, c=over_50.target, cmap="winter");
ax1.set(title="Heart disease for age & cholesterol level", xlabel='age', ylabel="chol")
ax1.set_xlim(45, 85)
ax1.axhline(y=over_50.chol.mean(), linestyle="--", color="red");
ax1.legend(*scatter.legend_elements(), title="Target")
ax1.set_xlim([50,80])
ax1.set_ylim([120,400])

scatter2 = ax2.scatter( x=over_50.age, y=over_50.thalach, c=over_50.chance, cmap="winter");
ax2.set_xlim(45, 85)
ax2.set(title="Heart disease for age & thalach",xlabel='age', ylabel="max heart rate");
ax2.legend(*scatter2.legend_elements(), title="Predicted heart disease chance");
ax2.set_xlim([50,80])

fig.suptitle("Heart Disease Analysis", fontsize=12, fontweight='bold');
```
